Hypothetis Testing

```python
#import the necessary library for this
from scipy.stats import ttest_ind
import scipy.stats as stats
```

```python
male=df[df['Sex']=='male']['Survived']
female=df[df['Sex']=='female']['Survived']

t_stat, p_value=ttest_ind(male, female)
print('T-statistics', t_stat)
print('P_Value', p_value)
```

```python
Adult=df[df['AgeGroup']=='Adult']['Survived']
Teen=df[df['AgeGroup']=='Teen']['Survived']
Child=df[df['AgeGroup']=='Child']['Survived']

f_stat, p_value=stats.f_oneway(Adult, Teen, Child)
print('F_Statistics', f_stat)
print('P_Value', p_value)
```

```python
#There is a statistically significant difference in survival rates among the different age groups (Child, Teen, and Adult).
```
